```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_name = 'mona.jpg'
```

```python
# 1: Выравнивание гистрограммы
lab_image = cv2.cvtColor(cv2.imread(img_name), cv2.COLOR_RGB2LAB)
L, A, B = cv2.split(lab_image)
```

```python
histL = cv2.calcHist([lab_image],[0],None,[256],[0,256])
plt.plot(histL)
```

```python
equalized_brightness = cv2.merge([cv2.equalizeHist(L), A, B])
plt.imshow(cv2.cvtColor(equalized_brightness, cv2.COLOR_LAB2RGB))
```

```python
histL = cv2.calcHist([equalized_brightness],[0],None,[256],[0,256])
plt.plot(histL)
```

```python
# 2: Размытие обычное
plt.imshow(cv2.cvtColor(cv2.blur(lab_image, (20, 20)), cv2.COLOR_LAB2RGB))
```

```python
# Гаусс
plt.imshow(cv2.cvtColor(cv2.GaussianBlur(lab_image, (21, 21), 0), cv2.COLOR_LAB2RGB))
```

```python
# Собель
sobel = cv2.Sobel(lab_image, cv2.CV_64F, 1, 1, ksize=7)
plt.imshow(sobel)
```

```python
# Лапрасс
laplas = cv2.Laplacian(lab_image, cv2.CV_64F)
plt.imshow(laplas)
```

```python
# Нормал
plt.imshow(cv2.cvtColor(cv2.medianBlur(lab_image, 21), cv2.COLOR_LAB2RGB))
```

```python
# Bilateral
plt.imshow((cv2.cvtColor(cv2.bilateralFilter(lab_image, 21, 40, 40), cv2.COLOR_LAB2RGB)))
```

```python
letter = cv2.imread('alph.jpg')
plt.imshow(letter)
```

```python
plt.imshow(cv2.morphologyEx(letter, cv2.MORPH_OPEN, np.ones(13, (np.random.random_integers(2, 5) + 1))))
```

```python
plt.imshow(cv2.morphologyEx(letter, cv2.MORPH_CLOSE, np.ones(13, (np.random.random_integers(2, 5) + 1))))
```
