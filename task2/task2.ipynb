```python
import math
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
file_name = "screen.jpg"
input_path = ""
output_path = ""
```

```python
# Task2: Увеличение насыщенности
def my_saturation(inp_path, out_path, image_name, light_koef):
    print("...")
    image = cv2.imread(inp_path + image_name)
    plt.imshow(image)
    lab_image = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
    l, a, b = cv2.split(lab_image)

    k, m = a.shape
    for i in range(k):
        for j in range(m):
            a[i, j] = min(255, a[i, j] * light_koef)
    k, m = b.shape
    for i in range(k):
        for j in range(m):
            b[i, j] = min(255, b[i, j] * light_koef)

    img_new = cv2.merge([l, a, b])
    img_new = cv2.cvtColor(img_new, cv2.COLOR_LAB2BGR)
    cv2.imwrite(out_path + "result_" + image_name, img_new)
    print("done")
```

```python
my_saturation(input_path, output_path, file_name, 1.1)
```
